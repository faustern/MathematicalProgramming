# The absolute value.. the easiest function in maths which make things the most confusing!

Define the function $f(x) = ||x|-1|$ and plot $f(x)$ and $f(f(x))$ and $f(f(f(x)))$

```python
def f(x):
    return sympy.Abs(sympy.Abs(x) - 1)
```

```python
sympy.plotting.plot(f(x), f(f(x)), f(f(f(x))), (x, -10*sympy.pi, 10*sympy.pi), legend = True)
```

## Calculus

Solve the equation $x^2 - 1 = 0$. We have $x^2 - 1 = (x+1)(x-1)$ so either $x+1=0$ or $x-1=0$, i.e, x = 1 or $x= -1$

```python
sympy.solvers.solve(x**2-1,x)
```

```python
sympy.plot(x**2-1,(x,-2,2))
```

Solve the equation $x^3-3x^2+2x-1=0$ and then confirm by plotting the equation...

```python
def f(x):
    return x**3 - 3*(x**2) + 2*x - 1

sympy.solvers.solve(f(x), x)
sympy.plot(f(x),(x,-2,4))
```

```python
sympy.real_roots(x**3-1,x)
```
