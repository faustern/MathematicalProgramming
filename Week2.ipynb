Exercise 3: Write the following function in Python

$$
f(x)=\left\{
\begin{array}{ll}
-x & \textrm{if $x<0$ }\\
x & \textrm{if $x\geq 0$}.
\end{array}\right.
$$

```python
x = int(input('enter a number:'))

if x < 0:
    print(-x)
else:
    print(x)
```

\begin{equation}
f(x)=\begin{cases}
-x, &\text{if $\lvert x \rvert <1$}\\
\sin(x), &\text{if $1\leq \lvert x \rvert  <2$}\\
\cos(x), &\text{otherwise.}
\end{cases}
\end{equation}

$|x|$ is called the absolute value of $x$ and the definition is the function we just wrote in Exercise 3

```python
import math
```

```python
x = float(input('enter a number:'))

if abs(x) < 1:
    print(-x)
elif 1 <= abs(x) < 2:
    print(math.sin(x))
else:
    print(math.cos(x))
```
