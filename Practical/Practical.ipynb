### Problem (a)

```python
def re(p):
    return(str(p)[: :-1])
```

```python
num = []
for p in range(1, 1000):
    prod = int(str(p) + str(re(p)) + str(p))
    if prod % 7 == 0 and prod % 11 == 0:
        num.append(p)
print(num)
```
